# 1. Diamantes: preços por tipo  de corte.


```python
import seaborn
import pandas
```

```python
seaborn.get_dataset_names()
```

```python
diamonds = seaborn.load_dataset('diamonds')
```

```python
display(diamonds)
```

```python
diamonds_cut_price = diamonds[['cut', 'price']].groupby('cut').agg('sum')
```

```python
display(diamonds_cut_price)
```

```python
diamonds_graphic = seaborn.relplot(data=diamonds_cut_price, x='cut', y='price', hue='price')
```
