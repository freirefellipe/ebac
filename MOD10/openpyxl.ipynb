```python
from openpyxl import load_workbook
```

```python
credito_xlsx = load_workbook(filename='credito.xlsx')
credito = credito_xlsx.active
header = next(credito.values) # listar os cabeçalhos.
print(header)
```

Extraindo uma coluna da tabela:

```python
produtos_indice = header.index('qtd_produtos') # index pega o número que indexa o cabeçalho desejado.

produtos = []

for produto in credito.values:
  if produto[produtos_indice] != 'qtd_produtos':
    produtos.append(produto[produtos_indice])
```

Outra forma, mais avançada, de fazer:

```python
produtos = [produto[produtos_indice] for produto in credito.values if produto[produtos_indice] != 'qtd_produtos'] # produtos é um conlchete com um laço for...in e uma condição if se quiser deixar o cabeçalho de fora
```

```python
print(produtos)
```
