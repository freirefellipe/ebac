# 1. Diamantes: pre√ßos por tipo  de corte.


```python
import seaborn

diamonds_ds = seaborn.load_dataset('diamonds')

diamonds_cut_price_ds = diamonds_ds[['cut', 'price']].groupby('cut').agg('sum')

diamonds_graphic = seaborn.relplot(data=diamonds_cut_price_ds, x='cut', y='price', hue='price')
```

*√â poss√≠vel inferir no gr√°fico, a partir do total da soma de cada tipo de corte, que o tipo de corte mais solicitado foi o corte ideal, com 74513487; seguido do Premium, com 63221498; sendo o tipo Fair menos solicitado, com 7017600.*

Abaixo, op√ß√µes de visualiza√ß√£o dos Datasets:

# 2. N√∫mero de passageiros em dezembro por ano.

```python
import seaborn

flights_ds = seaborn.load_dataset('flights')

flights_dec_ds = flights_ds.query('month == "Dec"')
```

```python
with seaborn.axes_style('whitegrid'):

  flights_dec_graphic = seaborn.barplot(data=flights_dec_ds, x='year', y='passengers', color='steelblue')

  flights_dec_graphic.set(title='Passageiros em dezembro ao longo dos anos', xlabel='Ano', ylabel='Passageiros')

  flights_dec_graphic.get_legend()
```

*A partir do gr√°fico acima, infere-se que os meses de dezembro, ao longo dos anos, obteve cresimento consider√°vel de vendas de passagens aereas, com pequena estagna√ß√£o nos anos de 1957 a 1958.*

Abaixo, op√ß√µes de visualiza√ß√£o dos Datasets:

```python
display(flights_ds)
```

```python
display(flights_dec_ds)
```

# 3. N√∫mero de passageiros por m√™s entre 1949 e 1959.

```python
import seaborn

flights_ds = seaborn.load_dataset('flights')

flights10years = flights_ds.query('1949 <= year <= 1959')

with seaborn.axes_style('whitegrid'):

  flights10years_graph = seaborn.lineplot(data=flights10years, x='year', y='passengers', hue='month')
```

*Depreende-se do gr√°fico acima que houve crescimento na demanda por passagens de avi√£o ao longo dos anos, em todos os meses, com maior signific√¢ncia nos meses em azul, que v√£o de junho a setembro, dessa vez exibindo estagna√ß√£o praticamente todos os meses, com exce√ß√£o de agosto e setembro, em 1957 e 1958.*
