# 2. **KDE com FacetGrid**
---
### O KDE exibe uma linha junto às barras, que indica a mediana, facilitando o entendimento. Incluindo o FacetGrid, é possível dividir os gráficos com mais uma variável além de *x* e *y*.

```python
import seaborn

titanic_ds = seaborn.load_dataset('titanic')[['age','survived','class']]

with seaborn.axes_style('whitegrid'):

  titanic_graph = seaborn.FacetGrid(data=titanic_ds, col='class', hue='survived')

  titanic_graph.map(seaborn.histplot, 'age', kde=True)
```

```python
display(titanic_ds)
```

# 3. **Box Plot**
---
### É um dos melhores gráficos para representar resumidamente distribuição de variável numérica, exibindo: *mediana*, *quartis* e *outliers*.

*OBS: - O box plot esconde a distribuição dos grupos na variável, caso existam.*

**Exemplo**: distribuição de idade por classe:

```python
import seaborn

titanic_ds = seaborn.load_dataset('titanic')

with seaborn.axes_style('whitegrid'):

  titanic_graph = seaborn.boxplot(x=titanic_ds['age'], y=titanic_ds['class']) # escolhe quais colunas vão ser base do gráfico, sem precisar restringir o Dataset.

  titanic_graph.set(title='Distribuição de idade', xlabel='Idade', ylabel='Classe')
```

# 4. **Correlação**: ***dispersão***
---
#### Correlação de proporção entre mais de uma variável numérica.
OBS: recomenda-se evidenciar grupos com uma terceira variável numérica.

```python
import seaborn

iris_ds = seaborn.load_dataset('iris')

iris_short = iris_ds[['sepal_length', 'petal_length', 'species']]
```

Proporção de comprimento de pétala por sépala:

```python
with seaborn.axes_style('whitegrid'):

  iris_graph = seaborn.scatterplot(data=iris_short, x='petal_length', y='sepal_length', hue='species', palette='pastel')

  iris_graph.set(title='Proporção do comprimeinto de pétala por sépala', xlabel='Pétala (cm)', ylabel='Sépala (cm)')

  iris_graph.get_legend().set_title('Espécie')
```
